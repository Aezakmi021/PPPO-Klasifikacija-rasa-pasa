```python
def matrix_Bin(labels):
    # Pronalaženje jedinstvenih vrednosti i broja pojavljivanja
    labels_unique, labels_counts = np.unique(labels, return_counts=True)

    # Mapiranje naziva rasa na indekse
    label_to_index = {label: i for i, label in enumerate(labels_unique)}

    # Kreiranje binarnih oznaka koristeći identitetsku matricu
    labels_indices = np.array([label_to_index[label] for label in labels])
    labels_bin = np.eye(len(labels_unique))[labels_indices]

    # Ispis broja podvarijabli (rasa) i oblika rezultujuće matrice
    print("Number of SubVariables:", labels_bin.shape[1])
    print("Shape:", labels_bin.shape)

    return labels_unique, labels_bin

# Pretvaranje labels u jednodimenzionalni niz
labels_flat = labels.flatten()

# Poziv funkcije
labels_name, labels_bin = matrix_Bin(labels_flat)
labels_bin[0:9]
```

# Definisanje konstanti i tezinskih faktora

```python
# Veličina slike u pikselima po svakoj dimenziji.
img_size = image_resize

# Broj boja kanala za slike.
num_channels = 3

# Dužina jednodimenzionalnog niza slika.
img_size_flat = img_size * img_size

# Oblik slike kao tuple dimenzija (visina, širina, broj kanala).
img_shape = (img_size, img_size, num_channels)

# Broj klasa koje se predviđaju.
num_classes = Nber_of_breeds

# Funkcija za inicijalizaciju težina. Generiše nove tenzore za težine koristeći
# normalnu distribuciju sa srednjom vrednošću 0 i standardnom devijacijom 0.05.
def new_weights(shape):
    return tf.Variable(tf.truncated_normal(shape, stddev=0.05))

# Funkcija za inicijalizaciju pristranosti. Generiše nove tenzore za pristranosti
# sa konstantnom vrednošću 0.05.
def new_biases(length):
    return tf.Variable(tf.constant(0.05, shape=[length]))
```

# Definisanje konvolucionog sloja

```python
def new_conv_layer(input,              # Prethodni sloj.
                   num_input_channels, # Broj kanala u prethodnom sloju.
                   filter_size,        # Širina i visina svakog filtera.
                   num_filters,        # Broj filtera.
                   use_pooling=True,   # Korišćenje maksimalnog agrumenata.
                   use_dropout=True):  # Korišćenje sloja isključivanja (dropout).

    # Oblik težina filtera za konvoluciju.
    # Ovaj format je određen TensorFlow API-jem.
    shape = [filter_size, filter_size, num_input_channels, num_filters]

    # Kreiranje novih težina, odnosno filtera sa datim oblikom.
    weights = new_weights(shape=shape)

    # Kreiranje novih pristranosti, jedna za svaki filter.
    biases = new_biases(length=num_filters)

    # Kreiranje TensorFlow operacije za konvoluciju.
    # Primetite da su koraci postavljeni na 1 u svim dimenzijama.
    # Prvi i poslednji korak moraju uvek biti 1,
    # jer je prvi za broj slike, a poslednji za ulazni kanal.
    # Ali, npr. koraci=[1, 2, 2, 1] bi značili da filter
    # pomeramo 2 piksela duž x- i y-ose slike.
    # Punjenje je postavljeno na 'SAME', što znači da se ulazna slika
    # dopunjuje nulama tako da je veličina izlaza ista.
    layer = tf.nn.conv2d(input=input,
                         filter=weights,
                         strides=[1, 1, 1, 1],
                         padding='SAME')

    # Dodavanje pristranosti rezultatima konvolucije.
    # Vrednost pristranosti se dodaje na svaki kanal filtera.
    layer += biases

    # Koristimo pooling za smanjenje rezolucije slike?
    if use_pooling:
        # Ovo je 2x2 maksimalno agregiranje, što znači da razmatramo
        # 2x2 prozore i biramo najveću vrednost u svakom prozoru.
        # Zatim se pomeramo 2 piksela do sledećeg prozora.
        layer = tf.nn.max_pool(value=layer,
                               ksize=[1, 2, 2, 1],
                               strides=[1, 2, 2, 1],
                               padding='SAME')

    # Rektifikacija linearnom jedinicom (ReLU).
    # Računa se max(x, 0) za svaki ulazni piksel x.
    # Ovo dodaje nešto nelinearnosti formuli i omogućava
    # učenje složenijih funkcija.
    layer = tf.nn.relu(layer)

    # Ako se koristi sloj isključivanja (dropout).
    if use_dropout:
        layer = tf.nn.dropout(layer, keep_prob_conv)

    # Napomena da se ReLU obično izvršava pre pooling-a,
    # ali pošto je relu(max_pool(x)) == max_pool(relu(x)) možemo
    # sačuvati 75% operacija relu-om maksimalnim agrumenata.

    # Vraćamo rezultirajući sloj i težine filtera
    # jer ćemo kasnije prikazati težine.
    return layer, weights
```

# Definisanje flatten sloja

```python
def flatten_layer(layer):
    # Dobijanje oblika ulaznog sloja.
    layer_shape = layer.get_shape()

    # Pretpostavljamo da je oblik ulaznog sloja:
    # layer_shape == [broj_slika, visina_slike, širina_slike, broj_kanala]

    # Broj karakteristika je: visina_slike * širina_slike * broj_kanala
    # Možemo koristiti funkciju iz TensorFlow-a za izračunavanje ovoga.
    num_features = layer_shape[1:4].num_elements()

    # Preoblikovanje sloja u [broj_slika, broj_karakteristika].
    # Napomena da smo samo postavili veličinu druge dimenzije
    # na broj_karakteristika i veličinu prve dimenzije na -1
    # što znači da se veličina u toj dimenziji izračunava
    # tako da ukupna veličina tenzora ostaje nepromenjena nakon preoblikovanja.
    layer_flat = tf.reshape(layer, [-1, num_features])

    # Oblik ravnanog sloja sada je:
    # [broj_slika, visina_slike * širina_slike * broj_kanala]

    # Vraćamo i ravni sloj i broj karakteristika.
    return layer_flat, num_features
```

# Definisanje potpuno povezanog sloja

```python
def new_fc_layer(input,          # Prethodni sloj.
                 num_inputs,     # Broj ulaza iz prethodnog sloja.
                 num_outputs,    # Broj izlaza.
                 use_relu=True,
                 use_dropout=True): # Koristiti ReLU aktivaciju

    # Kreiranje novih težina i pristranosti.
    weights = new_weights(shape=[num_inputs, num_outputs])
    biases = new_biases(length=num_outputs)

    # Izračunavanje sloja kao matematičke operacije množenja
    # ulaza i težina, a zatim dodavanje pristranosti.
    layer = tf.matmul(input, weights) + biases

    # Koristiti ReLU aktivaciju
    if use_relu:
        layer = tf.nn.relu(layer)

    # Primena dropout-a
    if use_dropout:
        layer = tf.nn.dropout(layer, keep_prob_fc)

    return layer
```

```python
# Placeholder za ulazne slike. Tip podataka je float32.
# Shape argument određuje oblik placeholder-a: [None, img_size, img_size, num_channels]
# None u prvom redu znači da možemo proslediti bilo koji broj slika.
# img_size x img_size x num_channels je dimenzija svake slike.
# name argument daje ime ovom placeholder-u.
x = tf.placeholder(tf.float32, shape=[None, img_size, img_size, num_channels], name='x')

# Transformacija ulaznih podataka x u oblik pogodan za obradu u neuronskoj mreži.
# Slike se reoblikuju u četvorodimenzionalni tenzor sa dimenzijama [-1, img_size, img_size, num_channels].
# -1 znači da se dimenzija automatski određuje na osnovu preostalih dimenzija i broja elemenata.
x_image = tf.reshape(x, [-1, img_size, img_size, num_channels])

# Placeholder za stvarne vrednosti ciljnih klasa. Tip podataka je float32.
# Shape argument određuje oblik placeholder-a: [None, num_classes]
# None znači da možemo proslediti bilo koji broj ciljnih klasa.
# num_classes označava broj mogućih klasa.
# name argument daje ime ovom placeholder-u.
y_true = tf.placeholder(tf.float32, shape=[None, num_classes], name='y_true')

# Indeksi stvarnih ciljnih klasa, korisno za evaluaciju performansi modela.
# Koristi se funkcija argmax za pronalaženje indeksa najveće vrednosti u svakom redu.
y_true_cls = tf.argmax(y_true, axis=1)

# Placeholder za verovatnoće zadržavanja neurona u potpuno povezanim slojevima.
# Ovaj placeholder se koristi za primenu dropout-a tokom obuke radi regularizacije.
# Keep probability može biti promenljiva koja određuje koliki je procenat neurona zadržan.
# Koristi se float32 tip podataka.
keep_prob_fc = tf.placeholder(tf.float32)

# Placeholder za verovatnoće zadržavanja neurona u konvolutivnim slojevima.
# Ovaj placeholder se takođe koristi za primenu dropout-a radi regularizacije.
# Koristi se float32 tip podataka.
keep_prob_conv = tf.placeholder(tf.float32)
```

```python
# Specifikacija veličine filtera i broja filtera za prvi konvolutivni sloj.
filter_size1 = 5          # Filteri konvolucije su dimenzija 5 x 5 piksela.
num_filters1 = 32         # Ukupno ima 32 ovakva filtera.

# Specifikacija veličine filtera i broja filtera za drugi konvolutivni sloj.
filter_size2 = 4          # Filteri konvolucije su dimenzija 4 x 4 piksela.
num_filters2 = 64         # Ukupno ima 64 ovakva filtera.

# Specifikacija veličine filtera i broja filtera za treći konvolutivni sloj.
filter_size3 = 3          # Filteri konvolucije su dimenzija 3 x 3 piksela.
num_filters3 = 128        # Ukupno ima 128 ovakvih filtera.

# Specifikacija veličine potpuno povezanog sloja.
fc_size = 500             # Broj neurona u potpuno povezanom sloju.

# Kreiranje prvog konvolutivnog sloja.
layer_conv1, weights_conv1 = \
    new_conv_layer(input=x_image,
                   num_input_channels=num_channels,
                   filter_size=filter_size1,
                   num_filters=num_filters1,
                   use_pooling=True,
                   use_dropout=False)

# Kreiranje drugog konvolutivnog sloja.
layer_conv2, weights_conv2 = \
    new_conv_layer(input=layer_conv1,
                   num_input_channels=num_filters1,
                   filter_size=filter_size2,
                   num_filters=num_filters2,
                   use_pooling=True,
                   use_dropout=False)

# Kreiranje trećeg konvolutivnog sloja.
layer_conv3, weights_conv3 = \
    new_conv_layer(input=layer_conv2,
                   num_input_channels=num_filters2,
                   filter_size=filter_size3,
                   num_filters=num_filters3,
                   use_pooling=True,
                   use_dropout=True)

# Ravnanje izlaza trećeg konvolutivnog sloja kako bi se ulaz dao potpuno povezanom sloju.
layer_flat, num_features = flatten_layer(layer_conv3)

# Kreiranje prvog potpuno povezanog sloja.
layer_fc1 = new_fc_layer(input=layer_flat,
                         num_inputs=num_features,
                         num_outputs=fc_size,
                         use_relu=True,
                         use_dropout=True)

# Kreiranje drugog potpuno povezanog sloja, ovaj put sa izlaznim brojem klasa.
layer_fc2 = new_fc_layer(input=layer_fc1,
                         num_inputs=fc_size,
                         num_outputs=num_classes,
                         use_relu=False,
                         use_dropout=False)

# Izlazni sloj za predikciju klasa koristeći softmax aktivaciju.
y_pred = tf.nn.softmax(layer_fc2)

# Izračunavanje prediktivnih klasa na osnovu verovatnoća klasa.
y_pred_cls = tf.argmax(y_pred, axis=1)
```

# Kreiranje skupova za treniranje

```python
# Veličina grupe koja će se koristiti tokom obuke.
batch_size = 50

# Funkcija koja vraća ukupno `num` slučajno izabranih uzoraka i njihove oznake.
def next_batch(num, data, labels):
    '''
    Return a total of `num` random samples and labels.
    '''
    # Generisanje niza indeksa uzoraka.
    idx = np.arange(0 , len(data))

    # Mešanje indeksa.
    np.random.shuffle(idx)

    # Izbor prvih `num` indeksa nakon mešanja.
    idx = idx[:num]

    # Izdvajanje slučajno izabranih uzoraka i oznaka.
    data_shuffle = [data[i] for i in idx]
    labels_shuffle = [labels[i] for i in idx]

    return np.asarray(data_shuffle), np.asarray(labels_shuffle)
```

```python
def optimize(num_iterations, X):
    global total_iterations

    # Početno vreme izvršavanja optimizacije.
    start_time = time.time()

    # Dictionary koji će sadržati gubitke (loss) tokom obuke.
    losses = {'train':[], 'validation':[]}

    # Petlja koja prolazi kroz zadati broj iteracija.
    for i in range(num_iterations):
        total_iterations += 1

        # Dobijanje grupe uzoraka za obuku.
        x_batch, y_true_batch = next_batch(batch_size, X_train, y_train)

        # Formiranje feed_dict-a za obuku.
        feed_dict_train = {x: x_batch,
                           y_true: y_true_batch,
                           keep_prob_conv : 0.3,
                           keep_prob_fc : 0.4}

        # Formiranje feed_dict-a za validaciju.
        feed_dict_validation = {x: X_validation,
                                y_true: y_validation,
                                keep_prob_conv : 1,
                                keep_prob_fc : 1}

        # Pokretanje optimizatora koristeći trenutnu grupu uzoraka.
        session.run(optimizer, feed_dict=feed_dict_train)

        # Izračunavanje tačnosti na trening skupu.
        acc_train = session.run(accuracy, feed_dict=feed_dict_train)

        # Izračunavanje tačnosti na validacionom skupu.
        acc_validation = session.run(accuracy, feed_dict=feed_dict_validation)

        # Dodavanje gubitaka u dictionary.
        losses['train'].append(acc_train)
        losses['validation'].append(acc_validation)

        # Ispisivanje statusa obuke na određenim iteracijama.
        if (total_iterations % X == 0) or (i ==(num_iterations -1)):
            msg = "Iteration: {0:>6}, Training Accuracy: {1:>6.1%}, Validation Accuracy: {2:>6.1%}"
            print(msg.format(total_iterations, acc_train, acc_validation))

    # Krajnje vreme izvršavanja optimizacije.
    end_time = time.time()

    # Razlika između početnog i krajnjeg vremena.
    time_dif = end_time - start_time

    # Ispisivanje vremena trajanja obuke.
    print("Time usage: " + str(timedelta(seconds=int(round(time_dif)))))

    # Plotovanje gubitaka tokom obuke.
    plt.plot(losses['train'], label='Training loss')
    plt.plot(losses['validation'], label='Validation loss')
    plt.legend()
    _ = plt.ylim()
```

```python
def plot_images(images, cls_true, cls_pred=None):
    # Provera da li su dužine slike, stvarnih i predviđenih klasa jednake.
    assert len(images) == len(cls_true) == 12

    # Kreiranje figure sa 4x3 podgrafika.
    fig, axes = plt.subplots(4, 3)

    # Podešavanje razmaka između podgrafika.
    fig.subplots_adjust(hspace=0.3, wspace=0.3)

    # Prolazak kroz sve podgrafike i prikazivanje slika.
    for i, ax in enumerate(axes.flat):
        # Prikazivanje slike.
        ax.imshow(images[i].reshape(img_shape), cmap='binary')

        # Prikazivanje stvarnih i predviđenih klasa.
        if cls_pred is None:
            xlabel = "True: {0}".format(cls_true[i])
        else:
            xlabel = "True: {0}, Pred: {1}".format(cls_true[i], cls_pred[i])

        # Postavljanje oznake klasa na x-osi.
        ax.set_xlabel(xlabel)

        # Uklanjanje oznaka sa grafika.
        ax.set_xticks([])
        ax.set_yticks([])

    # Osiguravanje da se grafik pravilno prikaže u jednoj ćeliji Notebook-a.
    plt.show()
```
